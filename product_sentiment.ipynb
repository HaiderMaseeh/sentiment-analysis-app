```python
import re
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

def clean_text(text):
    text = text.lower()  # lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # remove punctuation and numbers
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra spaces
    return text
```

```python
df['clean_text'] = df['Summary'].apply(clean_text)

```

```python
df=df[['clean_text','Sentiment']]
```

```python
import re
df['tokens'] = df['clean_text'].apply(lambda x: re.findall(r'\b\w+\b', x))
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

# Load stopwords
stop_words = set(stopwords.words('english'))

# Define words to keep (important for sentiment)
negation_words = {'not', 'no', 'nor', "don't", "didn't", "wasn't", "won't", "shouldn't", "couldn't"}

# Remove those from stopwords
filtered_stopwords = stop_words - negation_words

# Apply stopword removal
df['tokens'] = df['tokens'].apply(lambda tokens: [word for word in tokens if word.lower() not in filtered_stopwords])
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

df['tokens'] = df['tokens'].apply(lambda tokens: [
    lemmatizer.lemmatize(word, get_wordnet_pos(pos))
    for word, pos in pos_tag(tokens)
])
```

```python
df.head()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

# 1. Join tokens back into text for tokenizer
df['joined_text'] = df['tokens'].apply(lambda x: ' '.join(x))

# 2. Initialize and fit tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['joined_text'])

# 3. Convert text to sequences of numbers
df['sequences'] = tokenizer.texts_to_sequences(df['joined_text'])

# 4. (Optional) View vocabulary
word_index = tokenizer.word_index
print("Vocabulary size:", len(word_index))
print("Word Index example:", dict(list(word_index.items())[:10]))  # show first 10
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Automatically find max length in your data
max_len = max(df['sequences'].apply(len))

# Pad sequences (default: pre-padding with 0)
df['padded_sequences'] = list(pad_sequences(df['sequences'], maxlen=max_len, padding='post'))

print("Max sequence length:", max_len)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Features (X) and labels (y)
X = df['padded_sequences'].tolist()
y = df['Sentiment'].tolist()  # or df['onehot_label'] if using one-hot

# Train-test split
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
import numpy as np
# 4 Double-check for leakage
overlap = np.intersect1d([tuple(x) for x in X_train], [tuple(x) for x in X_val])
print("Overlapping samples after fix:", len(overlap))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout

vocab_size = 10000
embedding_dim = 128
maxlen = maxlen  # example
units = 128

model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=maxlen),
    LSTM(units, dropout=0.2, recurrent_dropout=0.2),
    Dense(3, activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.build(input_shape=(None, maxlen))

model.summary()  # will show ~1,411,971 params (with the values above)
```

```python
import numpy as np

X_train = np.array(X_train)
X_val   = np.array(X_val)
y_train = np.array(y_train)
y_val   = np.array(y_val)

```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_val   = le.transform(y_val)
```

```python

model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=5, batch_size=64)
```

```python
import numpy as np

# Suppose you already have the same tokenizer and label encoder from training
# tokenizer = Tokenizer(...)
# le = LabelEncoder(...)
# maxlen = <your sequence length>

def predict_sentiment(user_text):
    # Convert text to sequence of integers
    seq = tokenizer.texts_to_sequences([user_text])

    # Pad the sequence to match training length
    padded_seq = pad_sequences(seq, maxlen=maxlen, padding='post')

    # Predict
    prediction = model.predict(padded_seq)

    # Get the predicted class index
    pred_class_index = np.argmax(prediction, axis=1)[0]

    # Convert class index back to label
    pred_label = le.inverse_transform([pred_class_index])[0]

    return pred_label

# Example usage
user_input = input("Enter a product review: ")
print("Predicted Sentiment:", predict_sentiment(user_input))
```
